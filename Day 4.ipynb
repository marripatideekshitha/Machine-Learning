```python
import numpy as np
```

##### 1. Creating Matrices (2D Arrays)

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([[7, 8, 9],
              [1, 2, 3]])

print("Matrix A:\n", a)
print("\nMatrix B:\n", b)
```

##### 2. Basic Matrix Operations

```python
print("\nMatrix Addition:\n", a + b)
print("\nMatrix Subtraction:\n", a - b)
print("\nMatrix Multiplication (Element-wise):\n", a * b)
```

##### 3. Dot Product

```python

C = np.array([[1, 2],
              [3, 4],
              [5, 6]])

D = np.array([[1, 2, 3],
              [4, 5, 6]])

print("\nMatrix C:\n", C)
print("\nMatrix D:\n", D)

dot_product = np.dot(C, D)
print("\nDot Product (C x D):\n", dot_product)
```

##### 4. Matrix Transpose

```python
print("\nTranspose of A:\n", a.T)
print("\nTranspose of C:\n", b.T)
```

##### 5. Linear Algebra Functions

```python
square_matrix = np.array([[4, 2],
                          [3, 1]])

print("\nSquare Matrix:\n", square_matrix)
```

```python
# Determinant
det = np.linalg.det(square_matrix)
print("\nDeterminant:", det)
```

```python
inv = np.linalg.inv(square_matrix)
print("Inverse:\n", inv)
```

```python
# Eigenvalues & Eigenvectors
eigen_vals, eigen_vecs = np.linalg.eig(square_matrix)
print("\nEigenvalues:", eigen_vals)
print("Eigenvectors:\n", eigen_vecs)
```

##### 6. Solving Linear Equations
##### Example: 2x + 3y = 9, 4x + y = 13


```python
coeff_matrix = np.array([[2, 3],
                         [4, 1]])

constants = np.array([9, 13])

solution = np.linalg.solve(coeff_matrix, constants)
print("\nSolution of linear equations (x, y):", solution)
```
