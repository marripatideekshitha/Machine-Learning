```python
import numpy as np
```

##### 1. Creating basic NumPy arrays

```python
# 1D array
arr1 = np.array([1, 2, 3, 4])
print("1D Array:", arr1)
```

```python
# 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array:\n", arr2)
```

##### 2. Checking array properties

```python
print("\nShape:", arr2.shape)
print("Data Type:", arr1.dtype)
print("Number of Dimensions:", arr1.ndim)
print("Size (total elements):", arr1.size)
```

##### 3. Creating arrays using NumPy functions

```python
zeros_arr = np.zeros((3, 3))
print("\nArray of zeros:\n", zeros_arr)
```

```python
ones_arr = np.ones((3, 4))
print("\nArray of ones:\n", ones_arr)
```

```python
range_arr = np.arange(0, 10, 2)
print("\nArray with range:\n", range_arr)
```

```python
lin_arr = np.linspace(0, 1, 5)
print("\nArray with linspace:\n", lin_arr)
```

##### 4. Basic NumPy operations

```python
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print("\nAddition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

##### 5. Useful Math Functions

```python
print("\nMean:", np.mean(a))
print("Max:", np.max(a))
print("Min:", np.min(a))
print("Square Root:", np.sqrt(a))
print("Sum:", np.sum(a))
```
