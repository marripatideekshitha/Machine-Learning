### Day 3 - NumPy Math, Stats & Random Module (Code)

##### NumPy Math, Statistics & Random Module

```python
import numpy as np
```

##### 1. Mathematical Operations

```python
arr = np.array([1, 2, 3, 4, 5])
```

```python
print("Array:", arr)
print("Square:", np.square(arr))
print("Square Root:", np.sqrt(arr))
print("Exponential:", np.exp(arr))
```

##### 2. Statistical Functions

```python
stats_arr = np.array([10, 20, 20, 30, 40, 50])
```

```python
print("\nArray for statistics:", stats_arr)
print("Mean:", np.mean(stats_arr))
print("Median:", np.median(stats_arr))
print("Standard Deviation:", np.std(stats_arr))
print("Variance:", np.var(stats_arr))
print("Minimum value:", np.min(stats_arr))
print("Maximum value:", np.max(stats_arr))
```

##### 3. Aggregate Functions

```python
print("\nSum:", np.sum(stats_arr))
print("Cumulative Sum:", np.cumsum(stats_arr))
print("Cumulative Product:", np.cumprod(arr))
```

##### 4. NumPy Random Module

```python
# Random integer
rand_int = np.random.randint(1, 50, size=5)
print("\nRandom Integers:", rand_int)

# Random float numbers (0 to 1)
rand_float = np.random.rand(5)
print("Random Floats:", rand_float)

# Random normal distribution (mean=0, std=1)
rand_normal = np.random.randn(5)
print("Random Normal Distribution:", rand_normal)

# Random choice from a list
choices = np.random.choice([10, 20, 30, 40, 50], size=3)
print("Random Choice:", choices)
```

##### 5. Setting a random seed

```python
np.random.seed(42)
print("\nRandom numbers with seed (repeatable):", np.random.rand(5))
```
